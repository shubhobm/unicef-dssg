#Highest Granularity till Sub Region 2
### Sampling: Weekly
### Aggregation: Mean


```
df_country_stats = df.groupby(['country_region','country_region_code','sub_region_1','sub_region_2']).resample('7d',on='date').mean()
df_country_stats.head()
# for i in range(len(df_country_stats)):
#     print(df_country_stats,'\n')
```

```
df_country_stats.reset_index(inplace=True)
```

```
df_country_stats.head()
```

```
df_country_stats.to_csv('Global_Mobility_Report_Weekly_SUB2.csv')
```
